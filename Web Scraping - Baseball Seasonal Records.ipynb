```python
import pandas as pd
import numpy as np
from statistics import mean
```

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
```

```python
driver = webdriver.Chrome(chromedriver)
```

```python
df = pd.read_csv("baseball_output12.csv",converters={"Year":str})
#Created csv outline with URLs as last rows
```

```python
df.head() #(below output is after everything ran)
```

```python
count = 1099
while count < len(df):
    #BATTING PAGE FIRST
    driver.get(df["Batting_URL"][count])
    
    #Wins vs Losses
    record = driver.find_elements_by_tag_name('p')
    wins = record[2].text.split(" ")[1].split("-")[0]
    losses = record[2].text.split(" ")[1].split("-")[1]
    df["Wins"][count] = wins
    df["Losses"][count] = losses
    df["Win-Loss_%"][count] = int(wins)/int(losses)
    WebDriverWait(driver,3)
    
    #Total Batting Salary
    value_batting = driver.find_elements_by_xpath('//table[@id="players_value_batting"]/tfoot/tr[not(@class="thead")]')
    value_batting_string = value_batting[0].text
    batting_total_salary = int(value_batting_string.split(" ")[-1].replace("$","").replace(",",""))
    df["Total_Batting_Salary"][count] = batting_total_salary
    #Average Batting Age  
    df["Average_Batting_Age"][count]=batting_average_age = value_batting_string.split(" ")[2].replace("$","").replace(",","")
    WebDriverWait(driver,3)
    
    #Average Years Batting
    years_batting = driver.find_elements_by_xpath('//td[contains(@data-stat,"seasons")]')
    years_batting_list = []
    for item in years_batting:
        years_batting_list.append(item.text)
    val = 0
    while val < len(years_batting_list):
        years_batting_list[val] = int(years_batting_list[val])
        val +=1
    df["Average_Years_Batting"][count] = mean(years_batting_list)
    WebDriverWait(driver,3)
    
    #CHANGE DRIVER TO PITCHING PAGE
    driver.get(df["Pitching_URL"][count])
    value_pitching = driver.find_elements_by_xpath('//table[@id="players_value_pitching"]/tfoot/tr[not(@class="thead")]')
    value_pitching_string = value_pitching[0].text
    #Total Pitching Salary
    df["Total_Pitching_Salary"][count] = float(value_pitching_string.split(" ")[-1].replace("$","").replace(",",""))
    #Average Pitching Age
    df["Average_Pitching_Age"][count] = value_pitching_string.split(" ")[2].replace("$","").replace(",","")
    WebDriverWait(driver,3)

    #Average Years Pitching
    years_pitching = driver.find_elements_by_xpath('//td[contains(@data-stat,"seasons")]')
    years_pitching_list = []
    for item in years_pitching:
        years_pitching_list.append(item.text)
    new_val = 0
    while new_val < len(years_pitching_list):
        years_pitching_list[new_val] = int(years_pitching_list[new_val])
        new_val +=1
    df["Average_Years_Pitching"][count] = mean(years_pitching_list)
    WebDriverWait(driver,3)
    
    count+=1
```

```python
height_to_inches = {"5'5":65,
"5'6":66,
"5'7":67,
"5'8":68,
"5'9":69,
"5'10":70,
"5'11":71,
"6'0":72, 
"6'1":73,
"6'2":74,
"6'3":75,
"6'4":76,
"6'5":77,
"6'6":78,
"6'7":79,
"6'8":80,
"6'9":81,
"6'10":82,
"6'11":83}
```

```python
count = 1167
while count < len(df):    
    #CHANGE DRIVER TO ROSTER
    #Batting & pitching weights & heights
    driver.get(df["Roster_URL"][count])
    batting_height_list = []
    pitching_height_list = []
    batting_weight_list = []
    pitching_weight_list = []
    new_batting_height_list = []
    new_pitching_height_list = []
    roster_table = driver.find_element_by_xpath('//table[@id="appearances"]')
    WebDriverWait(driver,3)
    roster_lines = roster_table.find_elements_by_css_selector('tr')
    i = 1
    while i < len(roster_lines)-1:
        string_row = roster_lines[i].text.split(" ")
        age = string_row[2]
        if age.isdigit() == True:
            weight = string_row[9]
            height = str(string_row[7]+string_row[8]).replace("''","").replace("""""","").replace("\"","")
            games_pitched = string_row[18]
        #Account for players with "HOF", which shifts the data left
        else:
            age = string_row[3]
            if age.isdigit() == True:
                weight = string_row[10]
                height = str(string_row[8] + string_row[9]).replace("''","").replace("""""","").replace("\"","")
                games_pitched = string_row[19]
            #Account for players with extra symbols and letters in names
            else:
                weight = string_row[11]
                height = str(string_row[9] + string_row[10]).replace("''","").replace("""""","").replace("\"","")
                games_pitched = string_row[20]
        if int(games_pitched)>0:
            pitching_height_list.append(height)
            pitching_weight_list.append(weight)
            i+=1
        else:
            batting_height_list.append(height)
            batting_weight_list.append(weight)
            i+=1
    new_batting_height_list = [height_to_inches[x] for x in batting_height_list]
    new_pitching_height_list = [height_to_inches[x] for x in pitching_height_list]
    new_batting_weight_list = [int(x) for x in batting_weight_list]
    new_pitching_weight_list = [int(x) for x in pitching_weight_list]
    df["Average_Height_of_Pitcher"][count]=mean(new_pitching_height_list)
    df["Average_Height_of_Batter"][count]=mean(new_batting_height_list)
    df["Average_Weight_of_Pitcher"][count]=mean(new_pitching_weight_list)
    df["Average_Weight_of_Batter"][count]=mean(new_batting_weight_list)
    WebDriverWait(driver,3)
    count+=1
```

```python
df.to_csv('baseball_output13.csv')
```
